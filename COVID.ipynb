```python
df=data[['dateRep','cases','countriesAndTerritories']]
```

```python
df.dtypes
```

```python
df['Date']=pd.to_datetime(df['dateRep'],dayfirst=True)
```

```python
df
```

```python
df=df.drop(columns=['dateRep'])
```

```python
df['total cases']=df.sort_values('Date').groupby('countriesAndTerritories')['cases'].cumsum().sort_index()
```

```python
df=df[['Date','countriesAndTerritories','total cases']]
df=df.rename(columns={'countriesAndTerritories':'Pays'})
```

```python
df=pd.pivot_table(df,index=['Date'],columns=['Pays'],values=['total cases'])
```

```python
df.columns=[col[1] for col in df.columns]
```

```python
df=df.fillna(0).astype(int)
```

```python
df=df.drop(columns=['Cases_on_an_international_conveyance_Japan'])
```

```python
df.columns=[col.replace('_',' ') for col in df.columns]
```

```python
#prendre Top10 pays
country_reserved = set()
for index,row in df.iterrows():
    country_reserved|= set (row[row>0].sort_values(ascending=False).head(10).index)
df2=df[list(country_reserved)]  
```
